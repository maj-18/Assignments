# Species Count

```python
iris_data['Classification'].value_counts()
```

```python
plt.title('Species Count',fontsize=25)
sns.countplot(iris_data['Classification'])
```

# Comparison of sepal width and length

```python
plt.figure(figsize=(15,8))
plt.title('Comparison between various species based on sepal length and width',fontsize=25)
sns.scatterplot(iris_data['SL'],iris_data['SW'],hue =iris_data['Classification'],s=50)
```

# Comparison of petal length and width

```python
plt.figure(figsize=(15,8))
plt.title('Comparison between various species based on petal length and width',fontsize=25)
sns.scatterplot(iris_data['PL'], iris_data['PW'], hue = iris_data['Classification'], s= 50)
```

# Checking Correlation

```python
plt.figure(figsize=(10,11))
sns.heatmap(iris_data.corr(),annot=True)
plt.plot()
```

# Pairplots

```python
sns.pairplot(iris_data,hue="Classification",height=4)
```
